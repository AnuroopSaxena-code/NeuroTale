```python
import pandas as pd
df = pd.read_csv('../data/books.csv')
```

```python
df.head()
```

Now, we will clean this dataframe.

```python
df_clean = df[['title','authors','average_rating','ratings_count','original_publication_year']]
```

```python
df_clean = df_clean.rename(columns={'title':'book_title','authors':'book_author','average_rating':'rating','ratings_count':'num_ratings','original_publication_year':'year'})
df_clean.head()
```

```python
# Checking for missing data
print(df_clean.isnull().sum())
```

```python
df_clean = df_clean.dropna()
# Dropping rows with missing values
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
print(df_clean.shape)
```

```python
df_clean.head()
```

Now, we have a clean dataset by only using the useful columns, renaming them and by removing missing and duplicate entries.
